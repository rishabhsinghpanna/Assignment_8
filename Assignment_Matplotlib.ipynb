Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

```python
# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
# Create a figure and use subplot to arrange four plots in a 2x2 grid
plt.figure(figsize=(10, 8))
```

```python
# Plot 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1, marker='o', color='blue')
plt.title('Line 1')
plt.tight_layout()
plt.show()
```

```python
# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)
plt.plot(x2, y2, marker='s', color='green')
plt.title('Line 2')
plt.tight_layout()
plt.show()
```

```python
# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)
plt.plot(x3, y3, marker='^', color='orange')
plt.title('Line 3')
plt.tight_layout()
plt.show()
```

```python
# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)
plt.plot(x4, y4, marker='*', color='red')
plt.title('Line 4')
plt.tight_layout()
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

A box plot, also known as a whisker plot, is a graphical representation of the distribution of a dataset. It displays the summary statistics of a dataset, such as the minimum, first quartile (Q1), median, third quartile (Q3), and maximum. The plot consists of a rectangular "box" that represents the interquartile range (IQR) and "whiskers" that extend from the box to the minimum and maximum values. Outliers may also be displayed as individual points beyond the whiskers.

Box plots are useful for visually assessing the central tendency, spread, and skewness of a dataset, as well as identifying potential outliers.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
# Combine the data into a single list
data = [box1, box2]

# Create a box plot
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.title('Box Plot of box1 and box2')
plt.ylabel('Values')

# Show the plot
plt.show()
```
